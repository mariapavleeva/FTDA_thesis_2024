```
import numpy as np 
import pandas as pd
import os
import random
import pandas as pd
import numpy as np
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import random
from tqdm import tqdm
```

```
ads = os.listdir('/work/ad_images')
non_ads = os.listdir('/work/non_ad_images')
```

```
image_data = []

y = []
for filename in tqdm(ads):
    image_path = os.path.join('/work/ad_images', filename)
    img = load_img(image_path, target_size=(224, 224))
    img_array = img_to_array(img)
    img_array /= 255.0  
    image_data.append(img_array)
    y.append(1)
```

```
image_data, y = [], []

for filename in tqdm(non_ads):
    image_path = os.path.join('/work/non_ad_images', filename)
    img = load_img(image_path, target_size=(224, 224))
    img_array = img_to_array(img)
    img_array /= 255.0 
    image_data.append(img_array)
    y.append(0)
image_data = np.array(image_data)
```

```
import pickle
pickle.dump(image_data, open('image_data.pkl', 'wb'))
pickle.dump(y, open('y.pkl', 'wb'))
```
